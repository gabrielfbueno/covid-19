```python
# Importando as bibliotecas necessárias

import pandas as pd
import numpy as np
```

```python
# Importando dados de Abril

abril = pd.read_csv('abril.csv', encoding='UTF-8', delimiter=";")
abril.describe()
```

```python
# Limpando os dados

# Corrigindo valores negativos de abril

abril['casosNovosAbril'] = abril['casosNovosAbril'].abs()
abril['obitosNovosAbril'] = abril['obitosNovosAbril'].abs()

# removendo linhas sem valores não utilizadas, como o do total de estado entre outras
abril = abril.dropna()

# removendo colunas desnecessária
del abril['semanaEpi']

abril
```

```python
# Importando dados de Maio

maio = pd.read_csv('maio.csv', encoding='UTF-8', delimiter=";")
maio.describe()
```

```python
# Limpando os dados

# Corrigindo valores negativos de maio

maio['casosNovosMaio'] = maio['casosNovosMaio'].abs()
maio['obitosNovosMaio'] = maio['obitosNovosMaio'].abs()

# removendo linhas sem valores não utilizadas, como o do total de estado entre outras
maio = maio.dropna()

# removendo colunas desnecessária
del maio['semanaEpi']

maio
```

```python
# Importando dados de Junho

junho = pd.read_csv('junho.csv', encoding='UTF-8', delimiter=";")
junho.describe()
```

```python
# Limpando os dados

# Corrigindo valores negativos de junho

junho['casosNovosJunho'] = junho['casosNovosJunho'].abs()
junho['obitosNovosJunho'] = junho['obitosNovosJunho'].abs()

# removendo linhas sem valores não utilizadas, como o do total de estado entre outras
junho = junho.dropna()

# removendo colunas desnecessária
del junho['semanaEpi']

junho.describe()
```

```python
# Importando dados de Julho

julho = pd.read_csv('julho.csv', encoding='UTF-8', delimiter=";")
julho.describe()
```

```python
# Limpando os dados

# Corrigindo valores negativos de julho

julho['casosNovosJulho'] = julho['casosNovosJulho'].abs()
julho['obitosNovosJulho'] = julho['obitosNovosJulho'].abs()

# removendo linhas sem valores não utilizadas, como o do total de estado entre outras
julho = julho.dropna()

# removendo colunas desnecessária
del julho['semanaEpi']

julho
```

```python
# Importando dados da população

populacao = pd.read_csv('populacao.csv', encoding='UTF-8', delimiter=";")
populacao
```

```python
# importandando as coordenadas de todos munícipios

coordenadas = pd.read_csv('coordenadas.csv', encoding='UTF-8', delimiter=";")
coordenadas
```

```python
# removendo último número de munícipio (Planilha disponibilizada pelo GOV não possui este último número)

coordenadas['GEOCODIGO_MUNICIPIO'] = coordenadas['GEOCODIGO_MUNICIPIO'].astype(str).str[:-1]
coordenadas
```

```python
# unindo as duas tabelas através dos número de munícipios

AbrilMaio = pd.merge(abril, maio, left_on='codmun', right_on='codmun', how='inner')
AbrilMaioJunho = pd.merge(AbrilMaio, junho, left_on='codmun', right_on='codmun', how='inner')
AbrilMaioJunhoJulho = pd.merge(AbrilMaioJunho, julho, left_on='codmun', right_on='codmun', how='inner')
AbrilMaioJunhoJulho
```

```python
# unindo os dados de população

AbrilMaioJunhoJulhoPopulacao = pd.merge(AbrilMaioJunhoJulho, populacao, left_on='codmun', right_on='codmun', how='inner')
AbrilMaioJunhoJulhoPopulacao
```

```python
# convertando coluna com o código dos munícipios para string (para dar merge com coordenadas)

AbrilMaioJunhoJulhoPopulacao ['codmun'] = AbrilMaioJunhoJulhoPopulacao['codmun'].astype(str)
AbrilMaioJunhoJulhoPopulacao ['codmun'] = AbrilMaioJunhoJulhoPopulacao['codmun'].astype(str).str[:-2]
AbrilMaioJunhoJulhoPopulacao ['codmun']
```

```python
# unindo dados com as coordenadas de cada munícipio

dados = pd.merge(AbrilMaioJunhoJulhoPopulacao, coordenadas, left_on='codmun', right_on='GEOCODIGO_MUNICIPIO', how='inner')
dados
```

```python
# deletando colunas que não serão utilizadas

del dados['estado_y']
del dados['municipio_y']
del dados['data_y']
del dados['GEOCODIGO_MUNICIPIO']
del dados['NOME_MUNICIPIO']
del dados['data_x']
del dados['casosNovosAbril']
del dados['casosNovosMaio']
del dados['casosNovosJunho']
del dados['casosNovosJulho']
del dados['obitosNovosAbril']
del dados['obitosNovosMaio']
del dados['obitosNovosJunho']
del dados['obitosNovosJulho']

dados
```

```python
# substituindo vírgula por pontos das coordenadas

dados['LONGITUDE'] = dados['LONGITUDE'].str.replace(',', '.')
dados['LATITUDE']= dados['LATITUDE'].str.replace(',', '.')

dados
```

```python
# Criando um cópia pra realizar as modificações

covid = dados
covid
```

```python
# Calculando números de casos por mês

covid['CasosMaio'] = covid['casosAcumuladoMaio'] - covid['casosAcumuladoAbril']
covid['CasosJunho'] = covid['casosAcumuladoJunho'] - covid['casosAcumuladoMaio']
covid['CasosJulho'] = covid['casosAcumuladoJulho'] - covid['casosAcumuladoJunho']
covid
```

```python
# Calculando números de obitos por mês

covid['obitosMaio'] = covid['obitosAcumuladoMaio'] - covid['obitosAcumuladoAbril']
covid['obitosJunho'] = covid['obitosAcumuladoJunho'] - covid['obitosAcumuladoMaio']
covid['obitosJulho'] = covid['obitosAcumuladoJulho'] - covid['obitosAcumuladoJunho']
covid
```

```python
# removendo colunas duplicadas colunas duplicadas tb
covid = covid.loc[:,~covid.columns.duplicated()]

# deletando colunas que não serão utilizadas

del covid['casosAcumuladoAbril']
del covid['casosAcumuladoMaio']
del covid['casosAcumuladoJunho']
del covid['casosAcumuladoJulho']
del covid['obitosAcumuladoAbril']
del covid['obitosAcumuladoMaio']
del covid['obitosAcumuladoJunho']
del covid['obitosAcumuladoJulho']

covid
```

```python
# corrigindo valores que são negativos, definindo como 0

covid[covid['CasosMaio'] < 0] = 0
covid[covid['CasosJunho'] < 0] = 0
covid[covid['CasosJulho'] < 0] = 0
covid[covid['obitosMaio'] < 0] = 0
covid[covid['obitosJunho'] < 0] = 0
covid[covid['obitosJulho'] < 0] = 0
```

```python
# exportando dados

covid.to_csv('database.csv', sep=';', encoding='utf-8')
```
